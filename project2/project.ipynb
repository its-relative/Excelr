```python
import os

# Define the base directory containing the resumes
base_dir = 'resumes categorised'

# Initialize a dictionary to store the file counts for each category
file_counts = {}

# Walk through the directory structure
for category in os.listdir(base_dir):
    category_path = os.path.join(base_dir, category)
    if os.path.isdir(category_path):
        file_counts[category] = sum([len(files) for _, _, files in os.walk(category_path)])

# Print the file counts for each category
for category, count in file_counts.items():
    print(f'{category}: {count} files')

# Calculate the total number of files across all categories
total_files = sum(file_counts.values())
print(f'Total files across all categories: {total_files}')

```

```python
import os
import pandas as pd
import zipfile
import docx
import pypandoc
import PyPDF2

# Define the base directory containing the resumes zip file
base_zip = 'resumes categorised.zip'

# Initialize a list to store the data
data = []

def read_docx(file):
    doc = docx.Document(file)
    return '\n'.join([paragraph.text for paragraph in doc.paragraphs])

def read_doc(file):
    return pypandoc.convert_file(file, 'plain')

def read_pdf(file):
    reader = PyPDF2.PdfFileReader(file)
    content = []
    for i in range(reader.numPages):
        page = reader.getPage(i)
        content.append(page.extract_text())
    return '\n'.join(content)

# Open the zip file
with zipfile.ZipFile(base_zip, 'r') as zip_ref:
    # Iterate through each file in the zip archive
    for file_name in zip_ref.namelist():
        # Print the file name for debugging purposes
        print(f'Processing file: {file_name}')
        
        # Extract the category from the file path (assuming folder structure)
        category = os.path.dirname(file_name).split('/')[-1]
        
        # Check if the category is valid
        if category:
            # Handle different file types
            with zip_ref.open(file_name) as file:
                try:
                    if file_name.endswith('.docx'):
                        content = read_docx(file)
                    elif file_name.endswith('.doc'):
                        content = read_doc(file)
                    elif file_name.endswith('.pdf'):
                        content = read_pdf(file)
                    else:
                        print(f'Skipped file: {file_name} (unsupported file type)')
                        continue

                    data.append({'Resume': content, 'Category': category})
                    print(f'Read file {file_name} with category {category}')
                except Exception as e:
                    print(f'Failed to read file {file_name}: {e}')
        else:
            # Print a message if the file is skipped
            print(f'Skipped file: {file_name} (no category)')

# Check if any data was collected
if not data:
    print('No data was collected. Please check the zip file structure.')

# Create a DataFrame from the data if data is present
if data:
    df = pd.DataFrame(data)
    # Save the DataFrame to a CSV file
    df.to_csv('resumes_data.csv', index=False)
    print('Resumes data saved to CSV file successfully!')
else:
    print('No data to save to CSV file.')
    
```
