```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("50_Startups.csv")
```

```python
df = data.copy()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1:]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder
oe = OneHotEncoder()
```

```python
x_train = pd.get_dummies(x_train).astype(int)
x_test = pd.get_dummies(x_test).astype(int)
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
con_cols = ["R&D Spend","Administration","Marketing Spend"]
```

```python
x_train[con_cols] = sc.fit_transform(x_train[con_cols])
x_test[con_cols] = sc.transform(x_test[con_cols])
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
# y_pred.shape
y_test.shape
```

```python
check_df = pd.DataFrame({"pred":np.array(y_pred).reshape(-1),"actual":np.array(y_test).reshape(-1)})
```

```python
check_df
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
