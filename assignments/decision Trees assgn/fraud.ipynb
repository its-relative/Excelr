```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Fraud_check.csv")

df = data.copy()
```

```python
df
```

```python
new_risk = df[df["Taxable.Income"] <= 30000]
new_good = df[df["Taxable.Income"] > 30000]
```

```python
new_risk["Risky"] = "Risky"
```

```python
new_good["Risky"] = "Good"
```

```python
new_risk
```

```python
new_good 
```

```python
df = pd.concat([new_good,new_risk], axis= 0)
```

```python
df = pd.get_dummies(df).astype(int)
```

```python
df
```

```python
y = df.iloc[:,:1]
y
```

```python
x = df.iloc[:,1:]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
dc  = DecisionTreeRegressor(random_state=42,max_depth=3)
dc.fit(x_train,y_train)
```

```python
y_pred = dc.predict(x_test)
```

```python
check = pd.DataFrame({"pred":y_pred,"actual":np.array(y_test).reshape(-1)})
```

```python
check["pred"] = check["pred"].astype(int)
```

```python
check
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
