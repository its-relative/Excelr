```python
con_cols = [col for col in x.columns if x[col].dtype != "object" ]
```

```python
con_cols
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
x_train[con_cols] = sc.fit_transform(x_train[con_cols])
x_test[con_cols] = sc.transform(x_test[con_cols])
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit_transform(x_train) 
```

```python
pca.transform(x_test)
```

```python
from sklearn.cluster import KMeans
wss = []
for i in range(2,10):
    km = KMeans(n_clusters=i)
    km.fit(x_train.iloc[:,:3]) # Because in the question it has been asked to take only top 3 PCA's.
    wss.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(2,10),wss)
```

Yes the elbow can be clearly seen at 3rd iteration. Hence the optimal number of clusters is 3 which is determined by the elbow curve of KMeans.
