```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y_actual = np.array([2,4,6,8,10])

beta  = np.random.rand()

lr = 0.01

for _ in range(100):
    y_pred = beta*x
    loss = np.sum(np.square(y_actual - y_pred)) # loss
    loss_derivative = -2 * np.sum((y_actual - y_pred)*x)
    beta = beta - (lr*loss_derivative)
    # beta = beta - (lr*loss_)
    print("Coefficient is", beta)
# print("Coefficient is", beta)
```
