```python
import numpy as np
import matplotlib.pyplot as plt
```

Plotting a line with just two points

```python
xpoints = np.array([0,6])
ypoints = np.array([0,255]) # notice [xo,x1] and [y0,y1] 
plt.plot(xpoints,ypoints) # notice the order first the x points and then y points.
plt.show()
 
```

```python
xPoints = [0,6,10,15] # works fine with list too
yPoints = [0,255,290,300]
plt.plot(xPoints,yPoints)
plt.show()
```

Now lets just plot a line with some random points

```python
xPoints = np.array([0,6,10,15])
yPoints = np.array([0,255,290,300])
plt.plot(xPoints,yPoints)
plt.show()
```

In order to make a dotted or dashed line we can do that by:

```python
xPoints = np.array([0,6,10,15])
yPoints = np.array([0,255,290,300])
plt.plot(xPoints,yPoints, linestyle = "dotted")
plt.show()
```

```python
xPoints = np.array([0,6,10,15])
yPoints = np.array([0,255,290,300])
plt.plot(xPoints,yPoints,linestyle = "dashed")
plt.show()
```

Lets give x labels ,y labels and title to the graph 

```python
xPoints = np.array([0,6,10,15])
yPoints = np.array([0,255,290,300])
plt.plot(xPoints,yPoints)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line joining random points")
plt.show()
```

lets create a scatter plot with the same points

```python
xPoints = np.array([0,6,10,15])
yPoints = np.array([0,255,290,300])
plt.scatter(xPoints,yPoints)
plt.show()
```
