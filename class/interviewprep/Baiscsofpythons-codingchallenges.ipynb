```python
#write a program to find if given number is prime or not



def prime_finder(n):
    if n<=1:
        return "not prime"
    elif n==2 or n==3:
        return "prime"
    elif n>3:
        for i in range(2,int(n**0.5)+1):
            if n % i == 0:
                return "not prime"
            else:
                return "prime"
    
```

```python
prime_finder(3)
```

```python
import random

def generate_number():
     return random.randint(100,201),random.randint(100,201),random.randint(100,201)
    
def main():
    while True:
        num1,num2,num3=generate_number()
        prod=num1*num2*num3
        guess=int(input(f"what is the product of {num1},{num2} and {num3}?"))
        
        if guess==prod:
            print("correct")
        else:
            print("not correct") 
            print(f"correct answer is {prod}")
        
        play_again = input("Do you want to play again? (yes/no): ").strip().lower()
        
        if play_again != "yes":
            print("Thanks for playing!")
            break
            
if __name__ == "__main__":
    main()
    
```
