```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Mercedes.csv")
```

```python
df
```

```python
x = df.iloc[:,2:]
# x
y = df.iloc[:,1:2]
y
```

```python
con_cols = [col for col in x.columns if x[col].dtypes != "object"]
cat_cols = [col for col in x.columns if x[col].dtypes == "object"] 

# con_cols
```

```python
# df.isna().sum()[df.isna().any()]
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train[con_cols] = sc.fit_transform(x_train[con_cols])
x_test[con_cols] = sc.transform(x_test[con_cols])
```

```python
x_test[con_cols]
```

```python
from sklearn.preprocessing import OneHotEncoder
oe = OneHotEncoder(handle_unknown= "ignore")
```

```python
# oe_train = pd.DataFrame(oe.fit_transform(x_train[cat_cols]).toarray(),columns=x_train.,index= x_train.index)
```

```python
oe_train = oe.fit_transform(x_train).toarray()
```

```python
oe_train = pd.DataFrame(oe_train,index=x_train[cat_cols].index)
oe_test = pd.DataFrame(oe.transform(x_test).toarray(),index=x_test[con_cols].index)
```

```python
oe_test.shape
```

```python
train_final = pd.concat([x_train[con_cols],oe_train],axis=1)
test_final = pd.concat([x_test[con_cols],oe_test],axis=1)
```

```python
train_final
```

```python
test_final
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
```

```python
train_final.columns = train_final.columns.astype(str)
```

```python
New_pcaData = pca.fit_transform(train_final,y_train)
```

```python
test_final.shape
```

```python
y_test.shape
```

```python
test_final.columns = test_final.columns.astype(str)
test_f = pd.DataFrame(pca.transform(test_final))
```

```python
test_f = test_f.iloc[:,:207]
```

```python
pca_data = pd.DataFrame(New_pcaData,columns=[f'pca{i}' for i in range(New_pcaData.shape[1])])
```

```python
pca_data
```

```python
train_f = pca_data.iloc[:,:207]
from sklearn.ensemble import RandomForestRegressor

rr = RandomForestRegressor()
rr.fit(train_f,y_train)

```

```python

y_pred = rr.predict(test_f)
```
