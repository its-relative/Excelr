```python
data = pd.read_csv("Microsoft_data.csv")
df = data.copy()
df
```

```python
df.isna().sum()[df.isna().any()]
```

```python
con_cols = [col for col in df.columns if df[col].dtype != "object"]
```

```python
con_cols
```

```python
cat_cols = [col for col in df.columns if df[col].dtype == "object"]
```

```python
cat_cols
```

```python
mis = [col for col in df.isna().sum()[df.isna().any()].index]
```

```python
mis
```

```python
missing = {}
cat = []
con = []
for i in mis:
    if i in con_cols:
        con.append(i)
    elif i in cat_cols:
        cat.append(i)
for i in range(2):
    missing["cat"] = cat
    missing["con"] = con
```

```python
missing
```

```python
if len(missing["cat"])+len(missing["con"]) == len(df.isna().sum()[df.isna().any()]):
    print("yes",len(missing["cat"]),len(missing["con"]))
```

```python
catmiss_df = df[missing["cat"]]
conmiss_df = df[missing["con"]]
```
