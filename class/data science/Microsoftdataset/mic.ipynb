```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Microsoft_data.csv")
df = data.copy()
df
```

```python
# len(df.isna().any())
# len()
```

```python
# df.isna().sum()[df.isna().sum()[df.isna().any()] > (df.shape[0])*0.7]
dropL = (df.isna().sum()[df.isna().sum() > (df.shape[0])*0.2]).index
```

```python
dropL
```

```python
df.drop(columns=dropL,inplace=True)
```

```python
con_cols = [col for col in df.columns if df[col].dtype != "object"]
```

```python
con_cols
```

```python
cat_cols = [col for col in df.columns if df[col].dtype == "object"]
```

```python
cat_cols
```

```python
mis = [col for col in df.isna().sum()[df.isna().any()].index]
```

```python
mis
```

```python
missing = {}
cat = []
con = []
for i in mis:
    if i in con_cols:
        con.append(i)
    elif i in cat_cols:
        cat.append(i)
for i in range(2):
    missing["cat"] = cat
    missing["con"] = con
```

```python
missing
```

```python
if len(missing["cat"])+len(missing["con"]) == len(df.isna().sum()[df.isna().any()]):
    print("yes",len(missing["cat"]),len(missing["con"]))
```

```python
catmiss_df = df[missing["cat"]]
conmiss_df = df[missing["con"]]
```

```python
conmiss_df
```

```python
catmiss_df
```

```python
catmiss_df.describe()
```

```python
for catCol in catmiss_df.columns:
    df[catCol].fillna(df[catCol].mode()[0],inplace = True)
```

```python
for i in catmiss_df.columns:
    if i in df.isna().sum()[df.isna().any()]:
        print(i)
    else:
        print("all good") 
```
